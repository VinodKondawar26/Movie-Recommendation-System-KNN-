```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
movies=pd.read_csv("/content/movies.csv")
movies.head()
```

```python
ratings=pd.read_csv("/content/ratings.csv")
ratings.head()
```

```python
ratings=ratings.drop("timestamp",axis=1)
ratings.head()
```

```python
#pivot table
movies_users=ratings.pivot(index="movieId",columns='userId',values="rating").fillna(0)
movies_users.head()
```

```python
from scipy.sparse import csr_matrix
my_movies=csr_matrix(movies_users.values)
my_movies
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(metric="cosine",algorithm="brute",n_neighbors=6)
model.fit(my_movies)
#brute or auto =algorithm
```

```python
#COSINE SIMILARITY BETWEEN TWO VECTORS LIKE x1&x2 or x2&x3

# Compute the cosine similarity between users

from sklearn.metrics.pairwise import cosine_similarity
user_similarity = cosine_similarity(movies_users.fillna(0))

```

```python
# Function to make movie recommendations for a given user
def recommend_movies(movie_id, n_neighbors=5):
    _, indices = model.kneighbors(my_movies[movie_id - 1].reshape(1, -1), n_neighbors=n_neighbors + 1)
    recommended_movie_indices = indices.flatten()[1:]
    recommended_movies = movies[movies['movieId'].isin(recommended_movie_indices)]['title'].unique()
    return recommended_movies


movie_id = 6789
recommendations = recommend_movies(movie_id)
print(f"Top 5 movie recommendations for movieId {movie_id}:")
for movie in recommendations:
    print(movie)
```
